```python
import pandas as pd
```

```python
ruta = './bomberos/ActuacionesBomberos_'
fechas = ['2017', '2018', '2019', '2020', '2021', '2022', '2023']
dataframes = []

encoding='ISO-8859-1'

for fecha in fechas:
    df = pd.read_csv(ruta+fecha+'.csv', sep=';', skiprows=0, encoding=encoding)

    dataframes.append(df)
    if fecha == '2019':
        encoding='utf-8'


    
resultado = pd.concat(dataframes, ignore_index=True)
resultado.dropna(inplace=True)
resultado
```

```python
import requests
import pandas as pd

for numero in range(0, 119):
    url = 'https://datos.madrid.es/egob/catalogo/212616-'+ str(numero) +'-policia-estadisticas.xlsx'
    extension = '.xlsx'
    respuesta = requests.get(url)
    
    if respuesta.status_code == 404:
        url = 'https://datos.madrid.es/egob/catalogo/212616-'+ str(numero) +'-policia-estadisticas.xls'
        respuesta = requests.get(url)
        extension = '.xls'
    


    if respuesta.status_code == 200:

        with open('./policia/policia'+str(numero)+extension, 'wb') as archivo_local:
            archivo_local.write(respuesta.content)
```

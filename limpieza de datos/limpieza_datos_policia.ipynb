### Pruebas

```python
meses = ['enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio', 'agosto', 'septiembre','octubre', 'noviembre', 'diciembre']
years = ['2018', '2019', '2020', '2021', '2022']
ruta = './policia/'
```

```python
df = pd.read_excel('./policia/2017/actuaciones_policia_enero.xlsx', sheet_name='SEGURIDAD', skiprows=2)
df2 = pd.read_excel('./policia/2017/actuaciones_policia_enero.xlsx', sheet_name='DETENIDOS X DISTRITOS', skiprows=2)
df3 = pd.read_excel('./policia/2017/actuaciones_policia_enero.xlsx', sheet_name='ACCIDENTES', skiprows=2)
df4 = pd.read_excel('./policia/2017/actuaciones_policia_enero.xlsx', sheet_name='EXPEDIENTES OAC', skiprows=2)
df5 = pd.read_excel('./policia/2017/actuaciones_policia_enero.xlsx', sheet_name='VENTA AMBULANTE', skiprows=2)
df6 = pd.read_excel('./policia/2017/actuaciones_policia_enero.xlsx', sheet_name='CONSUMO ALCOHOL', skiprows=2)
df7 = pd.read_excel('./policia/2017/actuaciones_policia_enero.xlsx', sheet_name='LOCALES', skiprows=2)
df8 = pd.read_excel('./policia/2017/actuaciones_policia_enero.xlsx', sheet_name='AA.TT.', skiprows=2)
```

```python
# juntamos todas las hojas en un dataframe en base a la columna de distrito

df = pd.merge(df, df2, on='DISTRITOS', how='outer')
df = pd.merge(df, df3, on='DISTRITOS', how='outer')
df = pd.merge(df, df4, on='DISTRITOS', how='outer')
df = pd.merge(df, df5, on='DISTRITOS', how='outer')
df = pd.merge(df, df6, on='DISTRITOS', how='outer')
df = pd.merge(df, df7, on='DISTRITOS', how='outer')
df = pd.merge(df, df8, on='DISTRITOS', how='outer')
```

```python
# establecemos el mes en una columna con formato datetime
df['MES'] = '2017-01'
```

## 2017

2017 es un año especial, porque decidieron llamar a todas las cosas diferentes los muy desgraciados

```python
df_2017 = pd.DataFrame()

for mes in meses:
    df = pd.read_excel('./policia/2017/actuaciones_policia_' + mes + '.xlsx', sheet_name='SEGURIDAD', skiprows=2)
    df2 = pd.read_excel('./policia/2017/actuaciones_policia_' + mes + '.xlsx', sheet_name='DETENIDOS X DISTRITOS', skiprows=2)
    df3 = pd.read_excel('./policia/2017/actuaciones_policia_' + mes + '.xlsx', sheet_name='ACCIDENTES', skiprows=2)
    df4 = pd.read_excel('./policia/2017/actuaciones_policia_' + mes + '.xlsx', sheet_name='EXPEDIENTES OAC', skiprows=2)
    df5 = pd.read_excel('./policia/2017/actuaciones_policia_' + mes + '.xlsx', sheet_name='VENTA AMBULANTE', skiprows=2)
    df6 = pd.read_excel('./policia/2017/actuaciones_policia_' + mes + '.xlsx', sheet_name='CONSUMO ALCOHOL', skiprows=2)
    df7 = pd.read_excel('./policia/2017/actuaciones_policia_' + mes + '.xlsx', sheet_name='LOCALES', skiprows=2)
    df8 = pd.read_excel('./policia/2017/actuaciones_policia_' + mes + '.xlsx', sheet_name='AA.TT.', skiprows=2)

    df = pd.merge(df, df2, on='DISTRITOS', how='outer')
    df = pd.merge(df, df3, on='DISTRITOS', how='outer')
    df = pd.merge(df, df4, on='DISTRITOS', how='outer')
    df = pd.merge(df, df5, on='DISTRITOS', how='outer')
    df = pd.merge(df, df6, on='DISTRITOS', how='outer')
    df = pd.merge(df, df7, on='DISTRITOS', how='outer')
    df = pd.merge(df, df8, on='DISTRITOS', how='outer')

    df['AÑO'] = 2017
    df['MES'] = mes

    df_2017 = df_2017.append(df)
    df_2017 = df_2017[df_2017['DISTRITOS'] != 'TOTAL']

# renombramos las columnas para que tengan el mismo nombre que en los otros años

df_2017.rename(columns={'DISTRITOS': 'DISTRITOS', 
                        'RELACIONADAS CON LAS PERSONAS': 'RELACIONADAS CON LAS PERSONAS',
                        'RELACIONADAS CON EL PATRIMONIO': 'RELACIONADAS CON EL PATRIMONIO',
                        'POR TENENCIA DE ARMAS': 'POR TENENCIA DE ARMAS',
                        'POR TENENCIA DE DROGAS': 'POR TENENCIA DE DROGAS',
                        'POR CONSUMO DE DROGAS': 'POR CONSUMO DE DROGAS',
                        'DETENIDOS E INVESTIGADOS': 'PERSONAS DETENIDAS E INVESTIGADAS',
                        'CON HERIDOS': 'CON VICTIMAS',
                        'SIN HERIDOS': 'SIN VICTIMAS',
                        'EXPEDIENTES': 'EXPEDIENTES',
                        'PROPIEDAD INTELECTUAL E INDUSTRIAL': 'PROPIEDAD INTELECTUAL E INDUSTRIAL',
                        'INFRACCIONES ALIMENTARIAS': 'INFRACCIONES ALIMENTARIAS',
                        'ADULTOS': 'MAYORES DE EDAD',
                        'MENORES': 'MENORES DE 18 AÑOS',
                        'INSPECCIONES Y ACTUACIONES': 'INSPECCIONES Y ACTUACIONES',
                        'DENUNCIAS': 'DENUNCIAS',
                        'EXPEDIENTES INSTRUIDOS': 'EXPEDIENTES INSTRUIDOS',
                        'CONTACTOS CON CENTROS ESCOLARES': 'CONTACTOS CON CENTROS ESCOLARES'}, inplace=True)

df_2017.to_csv('./policia/actuaciones_policia_2017.csv', index=False, sep=';', encoding='ISO-8859-1')
```

```python
for year in years:

    df_total = pd.DataFrame()

    for mes in meses:
        df = pd.read_excel('./policia/'+year+'/actuaciones_policia_' + mes + '.xlsx', sheet_name='SEGURIDAD', skiprows=2)
        df2 = pd.read_excel('./policia/'+year+'/actuaciones_policia_' + mes + '.xlsx', sheet_name='PERS. DETENIDAS X DISTRITOS', skiprows=2)
        df3 = pd.read_excel('./policia/'+year+'/actuaciones_policia_' + mes + '.xlsx', sheet_name='ACCIDENTES', skiprows=2)
        df4 = pd.read_excel('./policia/'+year+'/actuaciones_policia_' + mes + '.xlsx', sheet_name='EXPEDIENTES OAC', skiprows=2)
        df5 = pd.read_excel('./policia/'+year+'/actuaciones_policia_' + mes + '.xlsx', sheet_name='VENTA AMBULANTE', skiprows=2)
        df6 = pd.read_excel('./policia/'+year+'/actuaciones_policia_' + mes + '.xlsx', sheet_name='CONSUMO ALCOHOL', skiprows=2)
        df7 = pd.read_excel('./policia/'+year+'/actuaciones_policia_' + mes + '.xlsx', sheet_name='LOCALES', skiprows=2)
        df8 = pd.read_excel('./policia/'+year+'/actuaciones_policia_' + mes + '.xlsx', sheet_name='AA.TT.', skiprows=2)

        df = pd.merge(df, df2, on='DISTRITOS', how='outer')
        df = pd.merge(df, df3, on='DISTRITOS', how='outer')
        df = pd.merge(df, df4, on='DISTRITOS', how='outer')
        df = pd.merge(df, df5, on='DISTRITOS', how='outer')
        df = pd.merge(df, df6, on='DISTRITOS', how='outer')
        df = pd.merge(df, df7, on='DISTRITOS', how='outer')
        df = pd.merge(df, df8, on='DISTRITOS', how='outer')

        df['AÑO'] = int(year)
        df['MES'] = mes

        df_total = df_total.append(df)
        df_total = df_total[df_total['DISTRITOS'] != 'TOTAL']

    df_total.to_csv('./policia/actuaciones_policia_'+year+'.csv', index=False, sep=';', encoding='ISO-8859-1')
```

## 2023

2023 también es un año un poco especial

```python
df_2023 = pd.DataFrame()

meses_2023 = ['enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio', 'agosto', 'septiembre','octubre']

for mes in meses_2023:
    df = pd.read_excel('./policia/2023/actuaciones_policia_' + mes + '.xlsx', sheet_name='SEGURIDAD', skiprows=2)
    df2 = pd.read_excel('./policia/2023/actuaciones_policia_' + mes + '.xlsx', sheet_name='PERS. DETENIDAS X DISTRITOS', skiprows=2)
    df3 = pd.read_excel('./policia/2023/actuaciones_policia_' + mes + '.xlsx', sheet_name='ACCIDENTES', skiprows=2)
    df4 = pd.read_excel('./policia/2023/actuaciones_policia_' + mes + '.xlsx', sheet_name='EXPEDIENTES OAC', skiprows=2)
    df5 = pd.read_excel('./policia/2023/actuaciones_policia_' + mes + '.xlsx', sheet_name='VENTA AMBULANTE', skiprows=2)
    df6 = pd.read_excel('./policia/2023/actuaciones_policia_' + mes + '.xlsx', sheet_name='CONSUMO ALCOHOL', skiprows=2)
    df7 = pd.read_excel('./policia/2023/actuaciones_policia_' + mes + '.xlsx', sheet_name='LOCALES', skiprows=2)
    df8 = pd.read_excel('./policia/2023/actuaciones_policia_' + mes + '.xlsx', sheet_name='AA.TT.', skiprows=2)

    df = pd.merge(df, df2, on='DISTRITOS', how='outer')
    df = pd.merge(df, df3, on='DISTRITOS', how='outer')
    df = pd.merge(df, df4, on='DISTRITOS', how='outer')
    df = pd.merge(df, df5, on='DISTRITOS', how='outer')
    df = pd.merge(df, df6, on='DISTRITOS', how='outer')
    df = pd.merge(df, df7, on='DISTRITOS', how='outer')
    df = pd.merge(df, df8, on='DISTRITOS', how='outer')

    df['AÑO'] = 2023
    df['MES'] = mes

    df_2023 = df_2023.append(df)
    df_2023 = df_2023[df_2023['DISTRITOS'] != 'TOTAL']
    df_2023 = df_2023[df_2023['DISTRITOS'] != '* DATOS INCOMPLETOS POR CAMBIO DE APLICACIÓN, DE SACPM A GEDIE,']
    df_2023 = df_2023[df_2023['DISTRITOS'] != 'NO TENIENDO ESTA ÚLTIMA EXPLOTACIÓN TODAVÍA']

df_2023.to_csv('./policia/actuaciones_policia_2023.csv', index=False, sep=';', encoding='ISO-8859-1')
```
